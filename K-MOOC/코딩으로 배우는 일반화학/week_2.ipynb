## 1. 열역학의 기초1: 이상기체방정식

### $PV = nRT$
- 실험적 관찰을 통해 얻어진 방정식
- P: 압력, V: 부피, T: 온도
- P, V : 보일의법칙, V,T : 샤를의 법칙
- $ R = \frac{PV}{nT} = 0.082\frac{L*atm}{mol*K}$ 
- $ P = \frac{nRT}{V}$
    - 온도의 감소에 따라 압력과 부피가 모두 감소하는 경향
    - 기체분자의 종류와 상관없이 이상기체방정식으로 거의 설명 가능

### 실제기체방정식
#### Compressibility factor
- $Z = \frac{PV}{nRT} = 1$ (이상기체방정식에서)
- 실제기체의 경우: 대기압에서 1과 유사, 기압이 백단위로 커질수록 점차 커짐

### 반데르발스 기체방정식
### $(P+a\frac{n^2}{V^2})(V-nb)$
- $a\frac{n^2}{V^2}$ : 분자간 인력으로 인한 압력값 수정
- $nb$ : 실제 기체 분자의 부피로 인한 부피값 수정

## 2. 열역학의 기초1: 반데발스 기체방정식

```python
# 기초 그래프 실습
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0, 2, 0.01)

plt.plot(x,np.sin(2*np.pi*x), 'b-' , label='sin(x)')
plt.plot(x,np.cos(2*np.pi*x), 'r--', label='cos(x)')

plt.legend()
plt.xlabel('x')
plt.ylabel('function')
plt.title('Comparison between sin and cos')

plt.show()
```

```python
# 이상기체 방정식
V = np.arange(10, 50, 1) # in L
n = 1 # in mol
R = 0.082057 # in L atm/(K mol)
T = 273.15 # in K
P = n*R*T/V

plt.plot(V,P, 'bo' , label='T=273.15K')

plt.xlabel('Volume(L)')
plt.ylabel('Pressure(atm)')
plt.title('Ideal Gas Equation')

plt.legend()
plt.show()
```

```python
# 온도(K)에 따른 이상기체방정식 그래프 양상

V = np.arange(10, 50, 1) # in L
n = 1 # in mol
R = 0.082057 # in L atm/(K mol)

T1 = 100 # in K
P1 = n*R*T1/V

T2 = 200 # in K
P2 = n*R*T2/V

T3 = 400 # in K
P3 = n*R*T3/V

plt.plot(V,P1, 'bo' , label='T=100K')
plt.plot(V,P2, 'g^' , label='T=200K')
plt.plot(V,P3, 'rv' , label='T=400K')

plt.xlabel('Volume(L)')
plt.ylabel('Pressure(atm)')
plt.title('Ideal Gas Equation')

plt.legend()
plt.show()
```
