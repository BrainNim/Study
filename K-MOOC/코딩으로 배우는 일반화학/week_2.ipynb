# 2주차: 열역학기초

## 1. 열역학의 기초1: 이상기체방정식

### $PV = nRT$
- 실험적 관찰을 통해 얻어진 방정식
- P: 압력, V: 부피, T: 온도
- P, V : 보일의법칙, V,T : 샤를의 법칙
- $ R = \frac{PV}{nT} = 0.082\frac{L*atm}{mol*K}$ 
- $ P = \frac{nRT}{V}$
    - 온도의 감소에 따라 압력과 부피가 모두 감소하는 경향
    - 기체분자의 종류와 상관없이 이상기체방정식으로 거의 설명 가능

### 실제기체방정식
#### Compressibility factor
- $Z = \frac{PV}{nRT} = 1$ (이상기체방정식에서)
- 실제기체의 경우: 대기압에서 1과 유사, 기압이 백단위로 커질수록 점차 커짐

### 반데르발스 기체방정식
### $(P+a\frac{n^2}{V^2})(V-nb)$
- $a\frac{n^2}{V^2}$ : 분자간 인력으로 인한 압력값 수정
- $nb$ : 실제 기체 분자의 부피로 인한 부피값 수정

## 2. 열역학의 기초1: 반데발스 기체방정식

```python
# 기초 그래프 실습
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0, 2, 0.01)

plt.plot(x,np.sin(2*np.pi*x), 'b-' , label='sin(x)')
plt.plot(x,np.cos(2*np.pi*x), 'r--', label='cos(x)')

plt.legend()
plt.xlabel('x')
plt.ylabel('function')
plt.title('Comparison between sin and cos')

plt.show()
```

```python
# 이상기체 방정식
V = np.arange(10, 50, 1) # in L
n = 1 # in mol
R = 0.082057 # in L atm/(K mol)
T = 273.15 # in K
P = n*R*T/V

plt.plot(V,P, 'bo' , label='T=273.15K')

plt.xlabel('Volume(L)')
plt.ylabel('Pressure(atm)')
plt.title('Ideal Gas Equation')

plt.legend()
plt.show()
```

```python
# 온도(K)에 따른 이상기체방정식 그래프 양상

V = np.arange(10, 50, 1) # in L
n = 1 # in mol
R = 0.082057 # in L atm/(K mol)

T1 = 100 # in K
P1 = n*R*T1/V

T2 = 200 # in K
P2 = n*R*T2/V

T3 = 400 # in K
P3 = n*R*T3/V

plt.plot(V,P1, 'bo' , label='T=100K')
plt.plot(V,P2, 'g^' , label='T=200K')
plt.plot(V,P3, 'rv' , label='T=400K')

plt.xlabel('Volume(L)')
plt.ylabel('Pressure(atm)')
plt.title('Ideal Gas Equation')

plt.legend()
plt.show()
```

선형의 경향이 나타나 대략 잘 따르고 있는것으로 볼 수 있음

```python
n = 1
R = 0.082057
T = 300

Z_ideal = [1] *14

plt.plot(P, P*V/(n*R*T),'bo', label = 'Air at 300K')
plt.plot(P, Z_ideal, 'r-', label='idal Gas')

plt.xlabel('P')
plt.ylabel('Z')
```

압력이 커질수록 Z값이 1에서 멀어짐 -> 이상기체방정식과의 오차가 커짐

$(P+a\frac{n^2}{V^2})(V-nb)$  
$P=\frac{nRT}{V-nb}-a\frac{n^2}{V^2}$

```python
# 반데르발스 기체방정식 그리기
V = np.arange(1, 50, 1) # in L
n = 1 # in mol
R = 0.082057 # in L atm/(K mol)
T = 300 # in K

a = 1.38 # Nitrogen, in L^2 / mol^2
b = 0.038 # Nitrogen, in L/mol

P_vdW = n*R*T/(V-n*b) - a*n**2/V**2
P_ideal = n*R*T/V

plt.plot(V,P_vdW, 'bo' , label='vdW eq of Air')
plt.plot(V,P_ideal, 'r-' , label='ideal Gas eq')

plt.xlabel('Volume(L)')
plt.ylabel('Pressure(atm)')
plt.title('vdW Gas Equation')

plt.legend()
plt.show()
```

주어진 부피, 압력에서는 이상기체와 실제기체가 매우 유사함

컴프레서빌리티 펙터 $Z = \frac{PV}{nRT}$ 를 통해 추가검증

```python
V = np.arange(0.05, 50, 0.01) # in L
n = 1 # in mol
R = 0.082057 # in L atm/(K mol)
T = 300 # in K

a = 1.38 # Nitrogen, in L^2 / mol^2
b = 0.038 # Nitrogen, in L/mol

P_vdW = n*R*T/(V-n*b) - a*n**2/V**2
Z_vdW = P_vdW*V/(n*R*T)
P_ideal = n*R*T/V
Z_ideal = P_ideal*V/(n*R*T)

plt.plot(P_vdW,Z_vdW, 'bo' , label='vdW eq of Air')
plt.plot(P_ideal,Z_ideal, 'r-' , label='ideal Gas eq')

plt.xlabel('Pressure(atm)')
plt.ylabel('Z')
plt.title('Compressibility Factor')

plt.legend()
plt.show()
```

## 실습과제
#### Hydrogen, Nitrogen의 반데르발스 상수를 기반으로 Compressibility Factor vs. pressure 그림을 그리고 비교


캡처출처([https://stachemi.tistory.com/31](https://stachemi.tistory.com/31))

Hydrogen(H2)
- a = 0.2444
- b = 0.02661

Nitrogen(N2)
- a = 1.390
- b = 0.03913

```python
n = 1 # in mol
R = 0.082057 # in L atm/(K mol)
T = 300 # in K

a_H2 = 0.2444
b_H2 = 0.02661

a_N2 = 1.390
b_N2 = 0.03913

V_H2 = np.arange(0.04, 50, 0.01)
P_H2 = n*R*T/(V_H2-n*b_H2) - a_H2*n**2/V_H2**2
Z_H2 = P_H2*V_H2/(n*R*T)

V_N2 = np.arange(0.05, 50, 0.01)
P_N2 = n*R*T/(V_N2-n*b_N2) - a_N2*n**2/V_N2**2
Z_N2 = P_N2*V_N2/(n*R*T)

plt.plot(P_H2,Z_H2, 'k-' , label='Hydrogen')
plt.plot(P_N2,Z_N2, 'g-' , label='Nigrogen')

plt.xlabel('Pressure(atm)')
plt.ylabel('Z')
plt.title('Hydrogen vs. Nitrogen')

plt.legend()
plt.show()
```
