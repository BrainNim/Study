# 양자역학과 분자구조1

## 분자오비탈 소개

### 원자 오비탈

### $ V(\vec{r}) = - \dfrac{1}{4\pi \epsilon_0} \dfrac{Ze^2}{|\vec{r}|}$
### $ \rightarrow \psi(\vec{r}) $


### 분자오비탈(수소분자)

### $ V(\vec{R_A},\vec{R_B},\vec{r}) = - \dfrac{e^2}{4\pi \epsilon_0} \big( \dfrac{1}{|\vec{r}-\vec{R_A}|} + \dfrac{1}{|\vec{r}-\vec{R_B}|} \big) \;\;$      (전자-원자핵)
### $ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; + \dfrac{e^2}{4\pi \epsilon_0} \dfrac{1}{|\vec{R_A}-\vec{R_B}|} \;\; $      (원자핵-원자핵)
### $ \rightarrow \psi(\vec{R_A},\vec{R_B},\vec{r}) $

but, 위와 같은 슈뢰딩거 방정식을 정확하게 풀 수 있는 방법은 없음  
따라서 근사값을 구함

### 근사값을 구하는 방법
- 원자핵 2개의 위치를 임의로 고정시킴(본 오펜하이머 근사, Born-Oppenheimer Approximation)
- 원자핵을 고정시키고 전자만 움직이는 것으로 가정
- 원자핵의 무게가 전자에 비해 훨씬 무거워 전자가 훨씬 빠르게 움직이기 때문에 근사치를 구하기 위한 매우 성공적인 방법

- 원자핵 2개를 고정시켰으므로 슈뢰딩거 방정식 역시 $r$만으로 정의할 수 있음

### $ V(\vec{r}) = - \dfrac{e^2}{4\pi \epsilon_0} \big( \dfrac{1}{|\vec{r}-\vec{R_A}|} + \dfrac{1}{|\vec{r}-\vec{R_B}|} \big) \;\;$     
($\vec{R_A}, \vec{R_A}$ 는 고정된 상수)

### $\big[ -\frac{h^2}{8m\pi^2}\Delta^2 + V(\vec{r}) \big]\Psi (\vec{r}) = E\Psi(\vec{r})$
- 일반적인 슈뢰딩거 방정식과 동일한 형태가 됨
