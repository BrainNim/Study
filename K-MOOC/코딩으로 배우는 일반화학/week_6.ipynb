# 6주차: 파동함수 소개

## 1. 파동함수 소개

### 파동
#### 파장 $\lambda$
#### 진폭
#### 반복되는 주기 : $\tau$
#### 주파수 : $\nu = \frac{1}{\tau}$
#### 파동의 속도 : $v = \lambda \times \nu$

### 2. 파동의 요소 간 관계 이해

#### 파동함수 : $ E(x, t) = E_{max}cos[2\pi (x/\lambda - vt)] $
- $x$:공간

#### 파동의 속도 : $v = \lambda \times \nu$
- 단, cos함수 내의 값이 변하지 않는다는 조건 하에 해당

```python
# 파동함수 애니메이션 만들기
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

# First set up the figure, the axis, and the plot element we want to animate
Lmax = 2.0  
Emax = 0.5  # 진폭
lamb = 1.0  # 파장
nu = 0.01  # 주파수

fig = plt.figure()
ax = plt.axes(xlim=(0, Lmax), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

# animation function.  This is called sequentially
def animate(i):
    x = np.linspace(0, Lmax, 1000)
    y = Emax * np.cos(2 * np.pi * (x/lamb - nu * i))
    line.set_data(x, y)
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=200, interval=20, blit=True)

# show animation
from IPython.display import HTML
HTML(anim.to_jshtml())
```
